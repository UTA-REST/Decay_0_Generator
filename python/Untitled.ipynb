```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.loadtxt("bb2nu_energy.txt")
len(y)
```

```python
bins=np.linspace(0,2.5,100)
#hist, bin_edges = np.histogram(y, bins=bins,density=True)
hist, bin_edges = np.histogram(y, bins=bins)
bin_centers = (bin_edges[1:] + bin_edges[:-1])/2
```

```python
plt.figure(figsize=(7,7))
plt.plot(bin_centers,hist,lw=3,color='k')

plt.ylabel(r"counts ",fontsize=22)
plt.xlabel("Energy [MeV]",fontsize=22)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.grid(which='both')
#plt.legend(loc='lower right',fontsize=17)

#plt.ylim(1e-3,1)
plt.xlim(0,2.5)
plt.tight_layout()
#plt.savefig("Townsstcomp.png",dpi=300)
plt.show()
```
